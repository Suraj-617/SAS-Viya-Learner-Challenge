# Splitting Data

```python
df.sampling.srs(
    samppct = 80,
    partind = True,
    seed    = 786,
    output  = dict(
        casOut = dict(name='Sampled',replace=True), 
        copyVars = 'ALL'),
    outputTables=dict(replace=True)
)
```

```python
sd=con.CASTable(name='Sampled')
```

# Checking Split id

```python
sd.hist(column="_PartInd_")
```

# Assigning common Parameters

```python
import collections 

params = dict(
    table    = dict(name = sd, where = '_partind_ = 1'), 
    target   = 'upsell_xsell', 
    inputs   = inputs, 
    nominals = 'upsell_xsell',
)

# Models to be trained
models = collections.OrderedDict()
#models['GBT'] = 'Gradient Boosting'
models['GBTune'] = 'Tuned Gradient Boosting'
```

# Train Gradient Boost

```python
con.decisionTree.gbtreeTrain( 
    **params, 
    lasso=1,
    ntrees=151,
    distribution="BINARY",
    maxBranch=3,
    ridge=1,
    subsamplerate=.8,
    seed = 786, 
    casOut = dict(name='GBTune', replace=True),
    savestate=dict(name='save_gbtune', replace=True),
    encodeName=True,
)
```

# Scorring Models

```python
def score_model(model):
    score = dict(
        encodeName=True,
        table      = 'Sampled',
        modelTable = dict(name='GBTune', caslib='CASUSER(suraj)'), # model + '_model',
        copyVars   = [target, '_partind_'],
        casOut     = dict(name = '_scored_' + model, replace = True)
    )
    return score

### Models Scorring
#con.decisionTree.gbtreeScore(**score_model('GBT'))
con.decisionTree.gbtreeScore(**score_model('GBTune'))
```

# Model Assessments

```python
def assess_model(model):
    assess = con.percentile.assess(
        table    = dict(name = '_scored_' + model, where = '_partind_ = 0'),
        inputs   = 'P_' + target + '1',      
        response = target,
        event    = '1',   
    )
    return assess

# Loop through the models and append to the roc_df dataframe
roc_df  = pd.DataFrame()
for i in range(len(models)):
    tmp = assess_model(list(models)[i])
    tmp.ROCInfo['Model'] = list(models.values())[i]
    roc_df = pd.concat([roc_df, tmp.ROCInfo])

# Display stacked confusion matrix using Python
print('\n', 'Confusion Matrix Information'.center(42, ' '))
roc_df[round(roc_df['CutOff'], 2) == 0.5][['Model', 'TP', 'FP', 'FN', 'TN']].reset_index(drop = True)
```

# Missclassification Rate

```python
# Add misclassification rate calculation
roc_df['Misclassification'] = 1 - roc_df['ACC']

print('\n', 'Misclassification Rate Comparison'.center(37, ' '))
miss = roc_df[round(roc_df['CutOff'], 2) == 0.5][['Model', 'Misclassification']].reset_index(drop = True)
miss.sort_values('Misclassification')
```

# ROC and ACU

```python
plt.figure(figsize = (7, 6))
for key, grp in roc_df.groupby(['Model']):
    plt.plot(grp['FPR'], grp['Sensitivity'], label = key + ' (C = %0.2f)' % grp['C'].mean())
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Postive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.title('ROC Curve (using validation data)');
```
